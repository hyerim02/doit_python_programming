## 3. 부가세 출력 프로그램  
---

소비자 가격 = 물건가격 $\times $ 1.1  
물건 가격 = 소비자 가격 $\times $ $\frac{1}{1.1}$  
부가세 = 물건 가격  $\times $ 0.1 = 소비자 가격 $\times $  $\frac{1}{11}$ 

```python
8000/1.1 #부가세 제외한 가격
```

```python
round(8000/1.1)
```

```python
8000-round(8000/1.1) #부가세
```

```python
round(8000 * (1/11)) #부가세
```

프로그래밍을 전문적으로 하려면 "협업"이 필수이기 때문에 `lambda`보다는 `def`를 더 사용해라!!

```python
price=[23, 40, 67]
for i in price:
    print (i * 1.1)
```

```python
def service_price():
    service = input('서비스 종류를 입력하세요, a/b/c: ')
    valueAdded = input('부가세를 포함합니까? y/n: ')
    if valueAdded == 'y':
        if service == 'a':
            result = 23 * 1.1
        if service == 'b':
            result = 40 * 1.1
        if service == 'c':
            result = 67 * 1.1 
    if valueAdded == 'n':
        if service == 'a':
            result = 23 
        if service == 'b':
            result = 40 
        if service == 'c':
            result = 67
    print (result)         
            
```

```python
service_price()
```
