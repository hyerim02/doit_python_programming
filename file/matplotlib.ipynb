데이터로 시각화 자료를 만드는 패키지  

데이터가 많을수록, 또 여러 그래프를 한 번에 만들어야 할 때 유용

```python
import matplotlib.pyplot as plt
```

```python
x=[1,4,9,16,25,36,49,64]
```

```python
plt.plot(x)
```

```python
plt.plot(x,color='deepskyblue')
```

```python
plt.plot(x,'or',color='cyan') # 점
```

```python
plt.plot(x,'-',color='skyblue') # 선
```

```python
plt.plot(x,'--',color='aqua') # 점선
```

```python
plt.plot(x,'^',color='lightskyblue') # 세모
```

```python
x
```

```python
y=[i for i in range(1,9)]
```

```python
y
```

```python
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('matplotlib sample')
```
